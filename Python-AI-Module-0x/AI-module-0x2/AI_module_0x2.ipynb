```python
#creating a dict 'company price' where key = car's company and value = median of their price

median_prices = CarData.groupby('CarCompany')['price'].median()

# Create a dictionary 'company_price' from the Series
company_price = median_prices.to_dict()

# Display the dictionary
sorted_keys = sorted(company_price.keys())

# Afficher les paires clé-valeur par ligne dans l'ordre alphabétique
for key in sorted_keys:
    print(f"{key}: {company_price[key]}")
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000
def map_price_bucket(price):
    if price < 10000:
        return 'Low'
    elif price >= 10000 and price <= 20000:
        return 'Medium'
    else:
        return 'High'

# Mapper les tranches de prix sur les entreprises de voitures
CarData['price'] = CarData['CarCompany'].map(lambda x: map_price_bucket(company_price[x]))

# Afficher les valeurs uniques de la colonne 'PriceCategory'
print(CarData['price'].unique())
CarData.CarCompany.unique()
```
